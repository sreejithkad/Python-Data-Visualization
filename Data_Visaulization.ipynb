# DATA VISUALIZATION ON TIPS DATASET, FOLLOWING PLOTS ARE USED

# 1) BAR CHART
# 2) HISTOGRAM CHART
# 3) PIE CHART
# 4) SWARM PLOT
# 5) JOINT PLOT
# 6) PAIR PLOT
# 7) VIOLIN PLOT
# 8) BOX PLOT
# 9) HEAT MAP

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv("/content/tips.csv")

data.head()
```

# ONE HOT ENCODING 

```python

data["sex"]=data["sex"].apply({'Male':1,'Female':0}.get)

data["smoker"]=data["smoker"].apply({'Yes':1,'No':0}.get)

data["day"]=data["day"].apply({'Thur':1,'Fri':2,'Sat':3,'Sun':4}.get)

data["time"]=data["time"].apply({'Dinner':1,'Lunch':2}.get)
```

```python
data.head()
```

# BAR CHART

```python
data.plot(kind='bar')
```

```python
data['total_bill'].plot(kind='bar')
```

```python
data['tip'].plot(kind='bar')
```

```python
data['size'].plot(kind='bar')
```

```python
data.head()
```

# HISTOGRAM CHART

```python
data['total_bill'].plot(kind='hist',stacked=True, bins=50)
```

```python
data['tip'].plot(kind='hist', stacked=True, bins=60)  # stacked= true means overlapping of histogram chart is somewhat avoided.
```

```python
data['size'].plot(kind='hist', stacked=False, bins= 20)  # when we decrease the bin size the width of histogram chart increases.
```

# SCATTER PLOT

```python
data.plot.scatter(x='total_bill', y='size')
```

# SWARM PLOT

```python
data.plot.hexbin(x='total_bill',y='tip', gridsize=5, C='time')   # X AXIS TALKS ABOUT TOTAL BILL, Y AXIS TALKS ABOUT TIP AND THE COLOUR INTENSITY TELLS ABOUT TIME
```

# JOINT PLOT

```python
snr.jointplot(x=data['total_bill'], y=data['tip'])
```

```python
snr.jointplot(x='total_bill', y='tip',data=data, hue='smoker')
```

```python
snr.jointplot(x=data['total_bill'], y=data['tip'], hue=data['sex'])
```

# PAIR PLOT

```python
snr.pairplot(data)
```

# VIOLIN PLOT

```python
snr.violinplot(x=data['time'], y=data['total_bill'])
```

```python
snr.violinplot(x=data['time'], y=data['tip'])
```
