# USING DATA VISUALIZATION ON IRIS DATASET.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
iris_data=snr.load_dataset("iris")
```

```python
iris_data.head()
```

```python
data=iris_data.drop(['species'],axis=1)

data.head()
```

```python
data.plot(figsize=(10,8), title='Iris Dataset')
```

# BAR CHART

```python
data.plot(kind='bar')
```

```python
data['sepal_length'].plot(kind='bar')
```

```python
data['sepal_width'].plot(kind='bar')
```

```python
data['petal_width'].plot(kind='bar')
```

# HISTOGRAM CHART

```python
data['sepal_width'].plot(kind='hist', stacked=True, bins=50 )
```

```python
data['sepal_length'].plot(kind='hist', stacked=True, bins=50)
```

# SCATTER PLOT

```python
data.plot.scatter(x='petal_width', y='petal_length')
```

```python
data.plot.scatter(x='sepal_length', y='sepal_width', c='petal_length')
```

# SWARM PLOT

```python
data.plot.hexbin(x='sepal_length', y='petal_width', gridsize=10, C='petal_length')
```

# JOINT PLOT

```python
snr.jointplot(x=data['sepal_length'], y=data['sepal_width'])
```

```python
snr.jointplot(x='sepal_length', y='petal_length', data=data, hue='sepal_width')
```

# PAIR PLOT

```python
snr.pairplot(data)
```

# PIE CHART

```python
x=data.iloc[0]

x.plot.pie(figsize=(5,5))
```

```python
x=data.iloc[1]

x.plot.pie(figsize=(10,10))
```

```python
x=data.iloc[0:5]

x.head()
```

```python
x.plot.pie(subplots=True, figsize=(15,15))
```

# VIOLIN PLOT

```python
snr.violinplot(x='sepal_length', y='petal_length', data=data)
```

# BOX PLOT

```python
snr.boxplot(x='sepal_length', y='petal_length', data=data)
```

# HEAT MAP

```python
snr.heatmap(data)
```
