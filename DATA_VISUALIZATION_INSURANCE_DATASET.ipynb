# DATA VISUALIZATION ON INSURANCE DATASET







```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv("/content/insurance.csv")

data.head()
```

```python
data.plot()
```

# BAR CHART

```python
data['bmi'].plot(kind='bar')
```

```python
data['charges'].plot(kind='bar')
```

# HISTOGRAM CHART

```python
data['bmi'].plot(kind='hist', stacked=True, bins=20)
```

```python
data['charges'].plot(kind='hist', stacked=False, bins=20)
```

# SCATTER PLOT

```python
data.plot.scatter(x='bmi', y='charges')
```

```python
data.plot.scatter(x='bmi', y='children')
```

```python
data.plot.scatter(x='bmi', y='charges', c='age')
```

# SWARM PLOT

```python
data.plot.hexbin(x='bmi', y='charges', gridsize=5, C='age')
```

# JOINT PLOT

```python
snr.jointplot(x=data['bmi'], y=data['charges'])
```

```python
snr.jointplot(x='bmi', y='charges', data=data, hue='age')
```

# PAIR PLOT

```python
snr.pairplot(data)
```

# VIOLIN PLOT

```python
snr.violinplot(x='smoker', y='bmi', data=data)
```

```python
snr.violinplot(x='sex', y='charges', data=data)
```

```python
snr.violinplot(x='sex', y='bmi', data=data)
```

# BOX PLOT

```python
snr.boxplot(x='sex', y='charges', data=data)
```

```python
snr.boxplot(x='sex', y='bmi', data=data)
```

# ONE HOT ENCODING

```python
data["sex"]=data["sex"].apply({'male':1, 'female':0}.get)

data["smoker"]=data["smoker"].apply({'yes':1,'no':0}.get)

data["region"]=data["region"].apply({'northeast':1, 'northwest':2, 'southeast':3, 'southwest':4}.get)
```
